```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convert_three(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_three)
```

```python
movies
```

```python
movies['crew'][0]
```

```python
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Directly inspect vocabulary_
vocabulary = cv.vocabulary_
print("Vocabulary:", vocabulary)

# Extract feature names from vocabulary
feature_names = sorted(vocabulary, key=vocabulary.get)
print("Feature names:", feature_names)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
import pandas as pd
from collections import defaultdict
import re

# Example data
data = new_df['tags'].tolist()

# Tokenization and vocabulary building
vocabulary = defaultdict(int)
for doc in data:
    tokens = re.findall(r'\b\w+\b', doc.lower())  # Basic tokenization
    for token in tokens:
        vocabulary[token] += 1

# Extract feature names from vocabulary (optional)
feature_names = list(vocabulary.keys())
print("Feature names:", feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
new_df['title'] = new_df['original_title']
new_df = new_df[['movie_id','title','tags']]
new_df
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
 
   
```

```python
recommend('Avatar')
```
